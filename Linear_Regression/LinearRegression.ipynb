## Imports: 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Generation of data:

```python
np.random.seed(42)

rooms = np.random.randint(1, 20, size=(50, 1))
prices = 50000 + 30000 * rooms + np.random.randint(0, 10000, size=(50, 1))

plt.scatter(rooms, prices)
plt.xlabel('Number of Rooms')
plt.ylabel('Price in USD')
plt.title('House Prices')
plt.show()
```

## Split data into training and testing sets:

```python
rooms_train, rooms_test, prices_train, prices_test = train_test_split(rooms, prices, test_size=0.2, random_state=42)
```

## Create and train the model:

```python
myModel = LinearRegression()

myModel.fit(rooms_train, prices_train)
```

## Evaluate the model:

```python
prices_pred = myModel.predict(rooms_test)

plt.scatter(rooms_test, prices_test, color='blue')
plt.plot(rooms_test, prices_pred, color='red')
plt.xlabel('Number of Rooms')
plt.ylabel('Price in USD')
plt.title('House Prices Prediction')
plt.show()
```
